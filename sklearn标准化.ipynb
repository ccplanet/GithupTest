```
# 0均值单位标准化方法 (X-mean)/std 
from sklearn import preprocessing 
import numpy as np  
X = np.array([[1., -1., 2.], [2., 0., 0.], [0., 1., -1.]])  
X_scaled = preprocessing.scale(X)  
# scaled之后的数据零均值，单位方差
ave = X_scaled.mean(axis=0)  # column mean: array([ 0.,  0.,  0.])  
std = X_scaled.std(axis=0)  #column standard deviation: array([ 1.,  1.,  1.])
print(X)
print(X_scaled)
print((X-X.mean(axis=0))/X.std(axis=0))
```

```
# 以下测试标准化的影响
# 导入包
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split # 训练测试数据集分割模块
from sklearn import preprocessing # 数据预处理模块
from sklearn.datasets.samples_generator import make_classification  # 用来生成分类数据
from sklearn.svm import SVC # 使用SVM模型训练
```

```
# 生成数据 绘图
#生成具有2个特征 300个样本数据
X, y = make_classification(
n_samples=300, n_features=2,
n_redundant=0, n_informative=2, 
random_state=20,  #种子
n_clusters_per_class=1, 
scale=100)
#可视化数据
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```
# 不做标准化训练
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
clf = SVC(gamma='auto')
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test)) # 准确率0.53
```

```
# 标准化后训练
X = preprocessing.scale(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
clf = SVC(gamma='auto')
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test)) # 准确率0.88
```
